__Stemming__

Stemming is the process of reducing a word to its base or root form, by removing any suffixes or prefixes that are attached to the word. The resulting stemmed word may not always be a valid word on its own, but it will represent the basic meaning of the original word.

```python
# importing Porterstemmer from nltk library
# checking for word 'research'

from nltk.stem import PorterStemmer

pst = PorterStemmer()
pst.stem('research')
```

```python
# checking for the list of words
stm = ['research', 'researching', 'researchers']

for word in stm:
    print(word + ':' + pst.stem(word))
```

__Lemmatisation__ 

In simpler terms, it is the process of converting a word to its base form. The difference between
stemming and lemmatization is, lemmatization considers the context and converts the word to its
meaningful base form, whereas stemming just removes the last few characters, often leading to incorrect meanings and spelling errors.

```python
# importing Lemmatizer library from nltk:
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

print('rocks:', lemmatizer.lemmatize('rocks'))
print('corpora:', lemmatizer.lemmatize('corpora'))
print('mining:', lemmatizer.lemmatize('mining'))
print('learning:', lemmatizer.lemmatize('learning'))
print('researching:', lemmatizer.lemmatize('researching'))
```

__Stop Words__

"Stop words" are the most common words in a language like "the", "a", "at", "for", "above", "on",
"is", "all". These words do not provide any meaning and are usually removed from texts. We can
remove these stop words using nltk library.

```python
# importing stopword form nltk library
from nltk import word_tokenize
from nltk.corpus import stopwords

a = set(stopwords.words('english'))

text = 'Cristiano Ronaldo was born on February 5, 1985, in Funchal, Madeira, Portugal.'
text1 = word_tokenize(text.lower())

print(text1)

stopwords = [x for x in text1 if x not in a]

print(stopwords)
```

__Part of speech tagging (POS)__

Part-of-speech tagging is used to assign parts of speech to each word of a given text (such as nouns,
verbs, pronouns, adverbs, conjunction, adjectives, interjection) based on its definition and its context. 

```python
text = 'vote to choose a particular student or a group (party) to represent them in Cergy University Conseil'

# Tokenize the text:
tex = word_tokenize(text)

for token in tex:
    print(nltk.pos_tag([token]))
```

__Named entity recognition__

It is the process of detecting the named entities such as the person name, the location name, the
company name, the quantities, and the monetary value.

```python
text = 'Cergy Paris University President Francois Germinet introduced the new research directives at Saint Martin campus'

# Importing chunk library from nltk:
import nltk
from nltk import ne_chunk
from nltk.tokenize import word_tokenize

# Tokenize and POS Tagging before doing chunk:
tokens = word_tokenize(text)
tags = nltk.pos_tag(tokens)
chunk = ne_chunk(tags)

print(chunk)
```

### Chunking
Chunking means picking up individual pieces of information and grouping them into bigger pieces.
In the context of NLP and text mining, chunking means a grouping of words or tokens into chunks.

```python
text = 'We saw the yellow dog'

token = word_tokenize(text)

tags = nltk.pos_tag(token)

reg = 'NP: {<DT>?<JJ>*<NN>}'

a = nltk.RegexpParser(reg)

result = a.parse(tags)
print(result)
```

```python
# import Wikipedia data
import wikipedia

articles = ['Data Science', 'Artificial intelligence', 'European Central Bank', 'Bank', 'Financial technology', 'International Monetary Fund', 'Basketball', 'Swimming']
# PageError: Page id "machine ;earning" does not match any pages. Try another id! - 'Machine Learning', - excluded
# PageError: Page id "tenic" does not match any pages. Try another id! - , 'Tennis' - excluded

wiki_lst = []
title = []

for article in articles:
    print("loading content: ", article)
    wiki_lst.append(wikipedia.page(article).content)
    title.append(article)

print("examine content")
print(wiki_lst)
```

```python
# import wiki data and represent each article as vector
import pandas as pd
import wikipedia
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

articles=['Data Science','Artificial intelligence','European Central Bank','Bank','Financial technology','International Monetary Fund','Basketball','Swimming']
wiki_lst=[]
title=[]
for article in articles:
    print("loading content: ",article)
    wiki_lst.append(wikipedia.page(article).content)
    title.append(article)

# vectorize the text data
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(wiki_lst)

# find the optimal number of clusters using the elbow method
silhouette_scores = []
k_range = range(2, 6)
for k in k_range:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

# plot the silhouette scores
plt.plot(k_range, silhouette_scores, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs Number of Clusters')
plt.show()

# cluster the data using k-means with k=3
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# print the cluster labels and corresponding article titles
labels = kmeans.labels_
for label, article in zip(labels, title):
    print(label, article)
```

```python
from wordcloud import WordCloud
result = {'cluster': labels, 'wiki': wiki_lst}
result = pd.DataFrame(result)
true_k = 3

for k in range(0, true_k):
    s = result[result.cluster == k]
    text = s['wiki'].str.cat(sep=' ')
    text = text.lower()
    text = ' '.join([word for word in text.split()])
    wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
    print('Cluster: {}'.format(k))
    titles = pd.Series(title)[result['cluster']==k].reset_index(drop=True)
    print('Titles:')
    print(titles.to_string(index=False))
    plt.figure()
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```
